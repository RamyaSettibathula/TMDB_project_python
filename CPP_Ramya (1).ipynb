### Identify missing values in the dataset and display the count of missing values

```python
df.isnull().sum()
```

```python
def seperate_data_types(df):
    categorical = []
    continous = []
    for column in df.columns:
        if df[column].dtypes == object:
            categorical.append(column)
        else:
            continous.append(column)
    return categorical,continous
categorical,continous = seperate_data_types(df)


from tabulate import tabulate
table = [categorical,continous]
print(tabulate({'Categorical': categorical,'Continous':continous}, headers =['categorical','continous']))
```

```python
def info_of_cat(col):
    print(f'Unique values in {col} are : {df[col].unique()}')
    print(f'mode of  {col} is : {df[col].mode()[0]}')
    print(f'total count of missing values in {col} is : {df[col].isnull().sum()}')
```

```python
def info_of_numerical(col):
    print(f'the mean of the {col} are : {df[col].mean()}')
    print(f'the median of the {col} are : {df[col].median()}')
    print(f'mode of  {col} is : {df[col].mode()[0]}')
    print(f'standard deviation of  {col} is : {df[col].std()}')
    print(f'total count of missing values in {col} is : {df[col].isnull().sum()}')
```

```python
info_of_cat('homepage')
```

```python
info_of_cat('tagline')
```

```python
df['tagline'].fillna('Based on a true story.', inplace = True)
```

```python
info_of_cat('release_date')
```

```python
df['release_date'].fillna('01-01-2006', inplace = True)
```

```python
info_of_cat('overview')
```

## Task 10: Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data: 
## 1. Budget
## 2. Revenue
## 3. Runtime
## Perform outlier analysis for the above three columns using box plots

```python
dsc = df.sort_values('revenue',ascending = False)
top500m = dsc.head(500)
top500m
```

```python
RBR = top500m[['budget','revenue','runtime']]
RBR.head()
```

```python
Mean = []
Median = []
Mode = []

for col in RBR:
    Median.append(RBR[col].median())
    Mean.append(RBR[col].mean())
    Mode.append(st.mode(RBR[col]))
    #Mode.append(salary_df[col].mode)

row_head = ['Mean', 'Median', 'Mode']
col_names = ['Budget','Revenue','Runtime']

# creating a data frame with all the calculated the mean, mode, median
data = pd.DataFrame([Mean,Median,Mode], columns = col_names)
data.insert(0,'Measures',row_head)
print('Central Tendency of the Budget,Revenue and Runtime')
data
```

```python
fig,ax = plt.subplots(1,2,figsize = (10,5))
sb.boxplot(data = RBR[['budget','revenue']],ax = ax[0])
sb.boxplot(data = RBR[['runtime']],ax = ax[1])
ax[0].set_title('Budget,Revenue before outliers analysis',fontsize=10)
ax[1].set_title('Runtime before outliers analysis')
plt.show()
```

```python
x = RBR[(RBR['revenue']>=6.585776e+08)].index
print(x)
for index in x:
    RBR.loc[index,'revenue'] = 4.585776e+08
```

```python
x = RBR[(RBR['budget']>=1.585776e+08)].index
print(x)
for index in x:
    RBR.loc[index,'budget'] = 103193736
```

```python
x = RBR[(RBR['runtime']>=175)].index
print(x)
for index in x:
    RBR.loc[index,'runtime'] = 118.72
```

```python
fig,ax = plt.subplots(1,2,figsize = (10,5))
sb.boxplot(data = RBR[['budget','revenue']],ax = ax[0])
sb.boxplot(data = RBR[['runtime']],ax = ax[1])
ax[0].set_title('Budget,Revenue after outliers analysis',fontsize=10)
ax[1].set_title('Runtime after outliers analysis')
plt.show()
```
